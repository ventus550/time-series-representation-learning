```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
from sklearn.manifold import TSNE
import datautils
from ts2vec import TS2Vec, tasks

train_data, train_labels, test_data, test_labels = datautils.load_arrowhead_data()
train_data = train_data[:, :, None]
test_data = test_data[:, :, None]
```

```python
train_data.shape
```

```python
model = TS2Vec(
    input_dims=1,
    device=0,
    output_dims=320
)
loss_log = model.fit(
    train_data,
    verbose=True
)
```

```python
tasks.eval_classification(model, train_data, train_labels, test_data, test_labels)
```
